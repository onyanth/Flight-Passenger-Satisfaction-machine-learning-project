```python
# start feature scaling
# split df into X and y
X=df.iloc[:,2:].values
y=df.satisfaction_v2.values
```

```python
X
```

```python
y
```

```python
#Splitting the data into training & test sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25)
```

```python
#Feature Scaling
from sklearn.preprocessing import MinMaxScaler
mms = MinMaxScaler()
X_train = mms.fit_transform(X_train)
X_test = mms.transform(X_test)

#fit_transform for training, and transform for test --> We fit the training data for model building
```

```python
X_train
```

```python
X_test
```

```python
# start trying different classifier models
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
model_kNN = KNeighborsClassifier()
model_dt = DecisionTreeClassifier(criterion='entropy')
model_rf = RandomForestClassifier(n_estimators=200)
model_nb = GaussianNB()
model_sv = SVC()
model_lr = LogisticRegression()

model_list = [model_kNN,model_dt,model_rf,model_nb,model_sv,model_lr]
```

```python
for model in model_list:
  model.fit(X_train, y_train)
  y_pred = model.predict(X_test)
  result = accuracy_score(y_test, y_pred)
  print(model,'\n','Accuracy is: ', result*100,'\n')
  print(classification_report(y_test, y_pred))
# test size: 0.25
```

```python
#Applying PCA
from sklearn.decomposition import PCA
```

```python
com_list = list(range(5,22,5))
result_list = []

for n_com in com_list:
  pca = PCA(n_components = n_com)
  X_train_pca = pca.fit_transform(X_train)
  X_test_pca = pca.transform(X_test)
  for model in model_list:
    model.fit(X_train_pca, y_train)
    y_pred = model.predict(X_test_pca)
    result = accuracy_score(y_test, y_pred)
    r = []
    r.append(n_com)
    r.append(model)
    r.append(result*100)
    result_list.append(r)

result_list
```

```python
# find out RandomForest generates the best accuracy, while PCA does not help saving a significant amount of time
```

```python
# so start resampling with SMOTEENN

from imblearn.combine import SMOTEENN
sm = SMOTEENN()
X_train_sen, y_train_sen = sm.fit_resample(X_train, y_train)

model_rf.fit(X_train_sen, y_train_sen)
y_pred_sen = model_rf.predict(X_test)

result_sen = accuracy_score(y_test, y_pred_sen)
print(model_rf,'\n','Accuracy is: ', result_sen*100,'\n')
print(confusion_matrix(y_test, y_pred_sen))
```

```python
print(classification_report(y_test, y_pred_sen))
```
